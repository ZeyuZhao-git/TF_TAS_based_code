```python
# 计算非线性聚合排名
import math

proxy_num = 0
s_rank = []

for i in range(len(af_list)):
    tmp = 0
    for k in rank_lists:
        considered_P = {'NASWOT','zen'} #'zen''te_nas''NASWOT''dss'
        if k in considered_P:
            tmp = tmp + math.log(rank_lists[k][i]/2)
            proxy_num = proxy_num + 1
        if k in {'dss', 'te_nas'}:
            tmp = tmp + math.log(rank_lists[k][i])
            proxy_num = proxy_num + 1
    #tmp = tmp - math.log(proxy_num) * proxy_num
    s_rank.append(tmp)
print(considered_P,s_rank)
```

```python
# 计算聚合排名相关系数

af_list = []
print(af_cand_info['af_all_res'])
for res in af_cand_info['af_all_res']:
    af_list.append(res['acc1'])
print('af_list:{}'.format(af_list))

#使用原始数据计算tau

kendall_s,_ = kendalltau(s_rank,af_list)

print('聚合排名相关系数:{}'.format(kendall_s))
```
